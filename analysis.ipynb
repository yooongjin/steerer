```python
from scipy import io
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import seaborn as sns
```

```python
UCF_QNRF_path = "/home/cho092871/Desktop/Datasets/UCF-QNRF/UCF-QNRF_ECCV18/Train"
```

```python
gt_files = glob.glob(os.path.join(UCF_QNRF_path, '*.mat'))
```

```python
gt_list = []
for file in gt_files:
    f = io.loadmat(file)
    gt_cnt = f['annPoints'].shape[0]
    gt_list.append(gt_cnt)
```

```python
max(gt_list), min(gt_list)
```

```python
sns.displot(np.array(gt_list), kde=True, bins = 15)
plt.title("UCF_QNRF Dist")
plt.ylabel("Num of data")
plt.xlabel("Num of Crowd")
plt.show()
```

```python
ucf = np.array(gt_list)
sns.displot(ucf[ucf < 2000], kde=True, bins = 15)
plt.title("UCF_QNRF Dist")
plt.ylabel("Num of data")
plt.xlabel("Num of Crowd")
plt.show()
```
